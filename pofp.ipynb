```python
from manim import *
from lambda_creature import LambdaCreature, ExpressionLambda

config.media_width = "100%"
config.verbosity = "WARNING"
config.quality = "medium_quality"
config.preview = False
```

```python

%%manim CreateCircle

class CreateCircle(Scene):
    def construct(self):
        circle = Circle()  # create a circle
        circle.set_fill(PINK, opacity=0.5)  # set the color and transparency
        self.play(Create(circle))  # show the circle on screen
```

```python

%%manim LambdaDemo

class LambdaDemo(Scene):
    def construct(self):
        # Create a simple lambda creature
        lambda_guy = LambdaCreature(body_color=BLUE_D, height=2.0)
        
        # Add to scene
        self.add(lambda_guy)
        
        # Fade in
        self.play(FadeIn(lambda_guy))
        self.wait(0.5)
        
        # Look at a point
        target = Dot(point=RIGHT * 2.5 + UP * 1.5, color=YELLOW, radius=0.15)
        self.play(FadeIn(target))
        self.play(ApplyFunction(lambda m: m.look_at(target.get_center()), lambda_guy), run_time=1)
        self.wait(1)
```

```python

%%manim NonDeterministicSort

class NonDeterministicSort(Scene):
    def construct(self):
        self.camera.background_color = BLACK
        true_color = ManimColor("#58C4DD")
        false_color = ManimColor("#FF644E")

        def boxed_number(value):
            box = RoundedRectangle(width=1.2, height=1.2, corner_radius=0.12)
            box.set_stroke(WHITE, width=2)
            num = MathTex(str(value))
            num.move_to(box.get_center())
            return VGroup(box, num)

        input_list = VGroup(
            boxed_number(2),
            boxed_number(1),
        ).arrange(RIGHT, buff=0.6)

        input_label = Tex("Input List").scale(0.7)
        input_label.next_to(input_list, DOWN, buff=0.35)

        self.play(
            FadeIn(input_list, shift=DOWN * 0.2),
            FadeIn(input_label, shift=DOWN * 0.2),
        )
        self.wait(0.3)

        coin_circle = Circle(radius=0.35, stroke_color=WHITE, stroke_width=2)
        coin_question = Tex("?").scale(0.9)
        coin = VGroup(coin_circle, coin_question)
        coin.next_to(input_list, UP, buff=0.6)
        self.play(FadeIn(coin, shift=UP * 0.3))
        self.wait(0.3)

        divider = Line(UP * 3.2, DOWN * 3.2)
        divider.set_stroke(WHITE, width=2)
        self.play(GrowFromCenter(divider))
        self.wait(0.2)

        left_list = input_list.copy().shift(LEFT * 3.2)
        right_list = input_list.copy().shift(RIGHT * 3.2)

        self.play(
            TransformFromCopy(input_list, left_list),
            TransformFromCopy(input_list, right_list),
        )
        self.play(FadeOut(input_list), FadeOut(input_label))

        left_label = Tex(r"if $2 < 1$ is True", color=true_color).scale(0.6)
        right_label = Tex(r"if $2 < 1$ is False", color=false_color).scale(0.6)
        left_label.next_to(left_list, UP, buff=0.4)
        right_label.next_to(right_list, UP, buff=0.4)

        self.play(
            FadeIn(left_label, shift=UP * 0.2),
            FadeIn(right_label, shift=UP * 0.2),
        )

        left_cell_a = left_list[0]
        left_cell_b = left_list[1]

        pos_a = left_cell_a.get_center()
        pos_b = left_cell_b.get_center()

        num_a = left_cell_a[1]
        num_b = left_cell_b[1]
        num_a_target = num_a.copy().move_to(pos_b)
        num_b_target = num_b.copy().move_to(pos_a)

        self.play(
            left_cell_a[0].animate.move_to(pos_b),
            left_cell_b[0].animate.move_to(pos_a),
            ReplacementTransform(num_a, num_a_target),
            ReplacementTransform(num_b, num_b_target),
            run_time=1.2,
        )
        left_cell_a.remove(num_a).add(num_a_target)
        left_cell_b.remove(num_b).add(num_b_target)
        self.wait(0.3)

        perm_a = Tex("Permutation A", color=true_color).scale(0.7)
        perm_b = Tex("Permutation B", color=false_color).scale(0.7)
        perm_a.next_to(left_list, DOWN, buff=0.6)
        perm_b.next_to(right_list, DOWN, buff=0.6)

        self.play(
            FadeIn(perm_a, shift=DOWN * 0.2),
            FadeIn(perm_b, shift=DOWN * 0.2),
        )
        self.wait(0.8)
```
